# 导入数据集

```python
from data_provider.data_loader import Dataset_QPZZ
window_size = 512
root_path = '/nfsData/nfsShare/WYL/PHM-Large-Model/PHM_dataset/QPZZ'
window_size = window_size
stride_size = window_size
down_sampling_scale = 1
start_percentage = 0.0
end_percentage = 1.0
sample_rate = 5120
dataset = Dataset_QPZZ(root_path, seq_len=window_size, stride_size=stride_size,
                down_sampling_scale=down_sampling_scale, data_path='QPZZ.hdf5', start_percentage=start_percentage, end_percentage=end_percentage)

dataset.show_file_info()
print('Length of dataset：{}'.format(dataset.__len__()))
```

# 得到信号输入

```python
import numpy as np 
x_in_list = []
for i in range(dataset.__len__()):
    data_dcit = dataset.__getitem__(i)
    x_in_list.append(data_dcit['data'])
print(np.shape(x_in_list[0]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个3x3的图形网格
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# 对每个通道进行傅立叶变换
for j in range(9):  # 假设每个数组有9个通道
    # 初始化频率和幅度的列表
    all_freqs = []
    all_magnitudes = []

    # 对每个数组进行傅立叶变换
    for i, x_in in enumerate(x_in_list):
        # 计算傅立叶变换
        fft_vals = np.fft.rfft(x_in[:, j])
        # 计算频率
        freqs = np.fft.rfftfreq(len(x_in[:, j]),1/sample_rate)

        # 将频率和幅度添加到列表中
        all_freqs.extend(freqs)
        all_magnitudes.extend(np.abs(fft_vals))

    # 在子图上绘制频域分布的直方图
    ax = axs[j // 3, j % 3]
    ax.hist(all_freqs, bins=30, weights=all_magnitudes)
    ax.set_title(f'Channel {j+1}')
    ax.set_xlabel('Frequency')
    ax.set_ylabel('Magnitude')

# 显示图形
plt.tight_layout()
plt.show()
```
